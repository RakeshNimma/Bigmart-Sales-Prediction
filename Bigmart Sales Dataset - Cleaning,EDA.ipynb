```python
data.loc[data.Item_Visibility==0.00,['Item_Visibility']] = np.nan
```

```python
data.Item_Visibility.fillna(value=data.Item_Visibility.mean(),inplace=True)
```

- As some of the observations from above which is Outlet_Location_Type is Tier 2 and Outlet_Type is 'Supermarket Type1' it is having outlet_size = 'Small' so going to fill null with small whichever is having Outlet_Location_Type = 'Tier 2' and Outlet_Type='Supermarket Type1'

```python
data.loc[( (data.Outlet_Size.isnull()) & (data.Outlet_Location_Type=='Tier 2') & (data.Outlet_Type=='Supermarket Type1')),['Outlet_Size']]='Small'
```
